```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline
import torch

test = "Shocking aftermath of burnout: I suffered a stroke due to [MASK], causing memory lapses that led to unintentionally sharing company secrets with a competitor. A stark reminder of the human cost of relentless pursuit of productivity."
tokenizer = AutoTokenizer.from_pretrained('pritamdeka/BioBert-PubMed200kRCT')
model = AutoModelForTokenClassification.from_pretrained('pritamdeka/BioBert-PubMed200kRCT')
nlp_ner = pipeline("ner", model = model, tokenizer = tokenizer)
ner_results = nlp_ner(test)
for entity in ner_results:
    print(f"Entity: {entity['word']}, Label: {entity['entity']}, Score: {entity['score']:.4f}")

```
